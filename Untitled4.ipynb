```python
from typing import List

import pandas as pd
import numpy as np
dataset = pd.read_csv('/content/age_salary_dataset.csv')
dataset
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print('Independent variable(X)',x)
print('Independent variable(Y)',y)
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan , strategy= 'mean')
x = imp.fit_transform(x)
y = y.reshape(-1,1)
y = imp.fit_transform(y)
y = y.reshape(-1)
print('mean of x:-',x)
print("mean of y:-",y)
x_df = pd.DataFrame(x,columns=dataset.columns[:-1])
y_df = pd.DataFrame(y,columns=[dataset.columns[-1]])
combined_data = pd.concat([x_df , y_df], axis=1)
print("Combined data:")
print(combined_data)

```

```python
combined_df = pd.DataFrame(combined_data, columns=['column1', 'column2', ...])
combined_df.to_excel('/content/age_salary_dataset.xlsx')
print('combined DATAFRAME :')
print(combined_df.head())
```
